```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv', encoding='euc-kr')
df.head()
```

```python
df[df['quantity']==3].head()
```

```python
df_quantity_3 = df[df['quantity']==3].head(5)
df_quantity_3.reset_index(drop=True)
```

```python
df_quantity_item_price = df[['quantity','item_price']]
df_quantity_item_price
```

```python
df['new_price'] = df['item_price'].str[1:].astype('float')
df['new_price'].head()
```

```python
# new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라
new_price_5 = len(df[df['new_price'] <= 5])
new_price_5
```

```python
# item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하고 index 값을 초기화 하여라

df_csb = df[df['item_name'] == 'Chicken Salad Bowl'].reset_index(drop=True)
df_csb.head(3)
```

```python
# new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라

df[(df['new_price'] <= 9) & (df['item_name'] == 'Chicken Salad Bowl')].head()
```

```python
# df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여라
df.sort_values('new_price').reset_index(drop=True).head()
```

```python
# df의 item_name 컬럼 값중 Chips 포함하는 경우의 데이터를 출력하라
df[df['item_name'].str.contains('Chips')]
```

```python
# df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라
df.iloc[:,::2]
```

```python
# df의 new_price 컬럼 값에 따라 내림차순으로 정리하고 index를 초기화 하여라

# help(df.sort_values)

df.sort_values(by='new_price', ascending=False).reset_index(drop=True)
```

```python
# df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 인덱싱하라
df_index = df[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
df_index
```

```python
# df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, 
# item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라
df_index = df[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
df_index.drop_duplicates(subset= 'item_name', keep='first')
```

```python
# df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, 
# item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라
df_index.drop_duplicates(subset = 'item_name', keep = 'last')
# help(df.drop_duplicates)
```

```python
# df의 데이터 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱하라
df[df['new_price'] >= df['new_price'].mean()]
```

```python
# df의 데이터 중 item_name의 값이 Izze 데이터를 Fizzy Lizzy로 수정하라
df.loc[df['item_name']=='Izze', 'item_name'] = 'Fizzy Lizzy'
df.head()
```

```python
# df의 데이터 중 choice_description 값이 NaN 인 데이터의 갯수를 구하여라
# df_Nan = df[df['choice_description'].isnull()]
# df_Nan = df.loc[df.choice_description.isnull()]
# len(df_Nan)
df.choice_description.isnull().sum()
```

```python
# df의 데이터 중 choice_description 값이 NaN 인 데이터를 NoData 값으로 대체하라(loc 이용)
df.loc[df.choice_description.isnull(), 'choice_description'] = 'NoData'
df.head()
```

```python
# df의 데이터 중 choice_description 값에 Black이 들어가는 경우를 인덱싱하라
ans = df.loc[df.choice_description.str.contains('Black')]
ans
# dir(df.choice_description.str)
```

```python
# df의 데이터 중 choice_description 값에 Vegetables 들어가지 않는 경우의 갯수를 출력하라
len(df.loc[~df.choice_description.str.contains('Vegetables')])
dir(df.choice_description.str)
```

```python
# df의 데이터 중 item_name 값이 N으로 시작하는 데이터를 모두 추출하라
df.loc[df.item_name.str.startswith('N')].head()
```

```python
# df의 데이터 중 item_name 값의 단어갯수가 15개 이상인 데이터를 인덱싱하라
# help(df.item_name.str.count)
df.loc[df.item_name.str.len() > 15].head()
```

```python
# df의 데이터 중 new_price값이 lst에 해당하는 경우의 데이터 프레임을 구하고 그 갯수를 출력하라
lst =[1.69, 2.39, 3.39, 4.45, 9.25, 10.98, 11.75, 16.98]
ans = df.loc[df.new_price.isin(lst)]
display(ans.head())
print(len(ans))
```
