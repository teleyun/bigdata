```python
# !git clone https://github.com/AnalyticsKnight/yemoonsaBigdata

import pandas as pd
data = pd.read_csv('housing_data.csv', header=None)
col_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV', 'isHighValue']
data.columns = col_names

data[data['MEDV'] > 36.9]
```

```python
data.info()
```

```python
data1 = data.copy()
data1['CRIM'] = data1['CRIM'].fillna(data['CRIM'].median())
```

```python
data2 = data.copy()
data2 = data.loc[data.CRIM.notnull(), ]
data2.describe()
data = data2
```

```python
Q1, Q3 = data['MEDV'].quantile([0.25, 0.75])
IQR = Q3 - Q1
upper_bound = Q3 + IQR * 1.5
lower_bound = Q1 - IQR * 1.5

outlier_upper = data['MEDV'] > upper_bound
outlier_lower = data['MEDV'] < lower_bound
outlier = outlier_upper.sum() + outlier_lower.sum()


print('outlier 범위 : %.2f 초과 또는 %.2f 미만' % (upper_bound, lower_bound))
print('outlier 개수 : %.0f 개' % outlier)
print('outlier 비율 : %.2f' % (outlier / len(data['MEDV'])*100) + '%')
```

```python
data4 = data.copy()
data4['MEDV'] = data4['MEDV'][(data4['MEDV'] <= upper_bound) & (data4['MEDV'] >= lower_bound)]
# (data['MEDV'] > 36.9).sum()

# data.loc[161:165, :]
```

```python
df_r = data.drop(['isHighValue'], axis = 1)
df_isHighvalue = data['isHighValue']
df_r[['MEDV', 'LSTAT', 'RM', 'CHAS', 'RAD', 'TAX']].corr()
```

```python
from sklearn.model_selection import train_test_split

X = df_r[['LSTAT', 'PTRATIO', 'TAX', 'AGE', 'NOX', 'INDUS', 'CRIM']]
y = df_r[['MEDV']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 123)

X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_minmax_scaled = scaler.transform(X_train)
X_test_minmax_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
model_lr.fit(X_train_minmax_scaled, y_train)
model_lr.coef_
```

```python
from sklearn.svm import SVR
model_svr = SVR()
model_svr.fit(X_train_minmax_scaled, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor(random_state=123)
model_rf.fit(X_train_minmax_scaled, y_train)
```
