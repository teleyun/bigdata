```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df.head(5)
```

```python
# 데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라

# ans = pd.DataFrame(df.groupby('host_name').size())
# ans = df.group_by('host_name')
ans = pd.DataFrame(df.host_name.value_counts().sort_index())
ans
# dir(df.host_name)
```

```python
# 데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라
ans = pd.DataFrame(df.host_name.value_counts().sort_values(ascending=False))
ans.rename(columns ={'host_name':'counts'}, inplace=True)
# ans['host_name'] = 'counts'
ans.head(5)
# help(df.host_name.value_counts().sort_values(ascending=False))
```

```python
# neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라
df.head()
# df[['neighbourhood_group', 'neighbourhood']].value_counts()
df.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size()
```

```python
# neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라
df_max = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index = False).size()
df_max.groupby('neighbourhood_group', as_index=False).max()
```

```python
# neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라
df[['neighbourhood_group', 'price']].groupby('neighbourhood_group', as_index = True).agg(['mean', 'var', 'max', 'min'])
```

```python
# neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라
# df.head()
df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min']))
df.groupby('neighbourhood_group').reviews_per_month.agg(['mean', 'var', 'max', 'min']))
```

```python
# neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 구하라
df.head()
# df[['neighbourhood', 'neighbourhood_group', 'price']].groupby(['neighbourhood', 'neighbourhood_group']).agg(['mean'])
df.groupby(['neighbourhood', 'neighbourhood_group']).price.mean()
```

```python
# neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하라
df.groupby(['neighbourhood', 'neighbourhood_group'], as_index=True).price.mean().unstack()
```

```python
# neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하고 nan 값은 -999값으로 채워라
df.groupby(['neighbourhood', 'neighbourhood_group']).price.agg(['mean']).unstack().fillna(-999)
# df.groupby(['neighbourhood', 'neighbourhood_group']).price.agg(['mean']).unstack().rename({' ':'-999'})
# dir(df)
```

```python
# 데이터중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값을 구하라
df[df.neighbourhood_group == 'Queens'].groupby('neighbourhood').price.agg(['mean', 'var', 'max', 'min']).head()
```

```python
# 데이터중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율을 구하여라
Ans = df.groupby(['neighbourhood_group', 'room_type']).size().to_frame().unstack()

Ans.iloc[:,:] = (Ans.values / Ans.sum(axis=1).values.reshape(-1,1))
Ans
```

```python
Ans = df[['neighbourhood_group','room_type']].groupby(['neighbourhood_group','room_type']).size().unstack()
Ans.iloc[:,:] = (Ans.values / Ans.sum(axis=1).values.reshape(-1,1))
Ans
```
