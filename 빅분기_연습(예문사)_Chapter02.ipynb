붓꽃 예측(p.172)

```python
from sklearn.datasets import load_iris
iris = load_iris()

iris
```

```python
import pandas as pd
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df['lable'] = iris.target
iris_df.head()

X = iris_df.loc[:,'sepal length (cm)':'petal width (cm)']
y = iris_df[['lable']]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 11)

print(len(X_train), len(X_test), len(y_train), len(y_test))

from sklearn.tree import DecisionTreeClassifier
df_clt = DecisionTreeClassifier(random_state  =11)
df_clt.fit(X_train, y_train)

pred = df_clt.predict(X_test)

dir(sklearn.metrics)
from sklearn.metrics import accuracy_score
ac_score = accuracy_score(y_test, pred)
ac_score
```

데이터 전처리(p180)

LabelEncoder 
- 숫자의 크기가 모형에 반영
- 선형회귀와 같은 알고리즘에는 부적합
- 트리 계열에 적합

```python
from sklearn.preprocessing import LabelEncoder

items = ['TV', '냉장고', '전자레인지', '컴퓨터','TV', '냉장고', '전자레인지', '컴퓨터' ]

encoder = LabelEncoder()

encoder.fit(items)

labels = encoder.transform(items)

print(labels)

# dir(encoder)
print(encoder.classes_)

items
```

원-핫 인코딩(sklearn)

```python
from sklearn.preprocessing import OneHotEncoder
# from sklearn import *
# dir(sklearn.linear_model)
import numpy as np
import pandas as pd

labels = labels.reshape(-1,1)
labels

oh_encoder = OneHotEncoder()
oh_labels = oh_encoder.fit(labels)
oh_labels = oh_encoder.transform(labels)

columns_ = encoder.classes_
type(oh_labels)
oh = oh_labels.toarray()
pd.DataFrame(oh, columns = columns_ ).astype('int')
```

원-핫 인코딩(getdummies)

```python
import pandas as pd

# dir(pd)
item_df = pd.DataFrame({'item' : items})
dummy = pd.get_dummies(item_df)

dummy
```

스케일링(정규화)

```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
iris_data = iris.data

iris_df = pd.DataFrame(iris_data, columns = iris.feature_names)

iris_df.head()

iris_df.head()

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(iris_df)
iris_std_scaled = scaler.transform(iris_df)
pd.DataFrame(iris_std_scaled, columns = iris.feature_names).head()
```

스케일링(최대-최소 정규화)

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(iris_df)
iris_df_minmax_scaled = scaler.transform(iris_df)

pd.DataFrame(iris_df_minmax_scaled, columns = iris.feature_names).agg(['min', 'max'])
```
